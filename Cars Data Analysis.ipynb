```python
import pandas as pd
```

```python
data = pd.read_csv('CarData.csv')
```

```python
data.head(10)
```

```python
data.shape
```

```python
## Data Cleaning- Find null values in dataset and fill it with mean() value
data.isnull()
```

```python
data.notnull()
```

```python
## count of data set having null values in columns
data.isnull().sum()
```

```python
data.info()
```

```python
data['Make']= data['Make'].fillna('Not Available')
data['Model']= data['Model'].fillna('Not Available')
data['Type']= data['Type'].fillna('Not Available')
data['Origin']= data['Origin'].fillna('Not Available')
data['DriveTrain']= data['DriveTrain'].fillna('Not Available')
data['MSRP']= data['MSRP'].fillna('Not Available')
data['Invoice']= data['Invoice'].fillna('Not Available')
```

```python
data.isna().sum()
```

```python
data['EngineSize']= data['EngineSize'].fillna(data['EngineSize'].mean())
data['Cylinders']= data['Cylinders'].fillna(data['Cylinders'].mean())
data['Horsepower']= data['Horsepower'].fillna(data['Horsepower'].mean())
data['MPG_City']= data['MPG_City'].fillna(data['MPG_City'].mean())
data['MPG_Highway']= data['MPG_Highway'].fillna(data['MPG_Highway'].mean())
data['Weight']= data['Weight'].fillna(data['Weight'].mean())
data['Wheelbase']= data['Wheelbase'].fillna(data['Wheelbase'].mean())
data['Length']= data['Length'].fillna(data['Length'].mean())
```

```python
data.head()
```

```python
## Check what are the different type of 'Make' are there is our dataset. what is the count of each make in data
data['Make'].value_counts().sort_values(ascending=False)
```

```python
## Show all data where Origin is 'Asia' or 'Europe'
data[(data['Origin'] == 'Asia') | (data['Origin'] == 'Europe')]
```

```python
data[data['Origin'].isin(['Asia', 'Europe'])]
```

```python
## Remove all records where weight is above 4000
data[~(data['Weight'] > 4000)]
```
