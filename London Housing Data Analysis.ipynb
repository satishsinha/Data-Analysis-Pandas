```python
data['no_of_crimes'] = data['no_of_crimes'].fillna('Not Available')
```

```python
data.head()
```

```python
data.info()
```

```python
## Convert date Datatype to Date Time format
data['date'] = pd.to_datetime(data.date)
data
```

```python
## Add new column 'month' as 2nd column in the dataframe, which contain month only
data.insert(1, 'months', data.date.dt.month)
data
```

```python
## Add a new column 'year' in the dataframe, which contain only year value
data.insert(2, 'year', data.date.dt.year)
data
```

```python
## Remove month and year from data frame
data.drop(['year', 'months'], axis=1, inplace = True)
data
```

```python
## Show all the records where 'no_of_crimes' is 0. 
zero_crime = data[data['no_of_crimes'] == 0]
zero_crime
```

```python
## How many records are there for above condition?
zero_crime.shape[0]
```

```python
data['year'] = data.date.dt.year
```

```python
## What is the minimum and maximum 'average_price' per year in england
data.groupby('year').average_price.mean()
```

```python
data.groupby('year').average_price.max()
```

```python
## What is the minimum and maximum 'No. of crimes' recoreded per area in england
df = data[~(data['no_of_crimes'] == 'Not Available')]
df.groupby(['area']).no_of_crimes.min().sort_values(ascending=False)
```

```python
df.groupby(['area', 'year']).no_of_crimes.min().sort_values(ascending=False)
```

```python
df.groupby('area').no_of_crimes.max().sort_values(ascending=False)
```

```python
df.groupby(['area', 'year']).no_of_crimes.max().sort_values(ascending=False)
```
