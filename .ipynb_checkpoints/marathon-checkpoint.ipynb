```python
## 1. Correlation Between Athlete Age and Average Speed
## What is the correlation between an athlete's age and their average speed?
## Does age significantly affect the speed, and if so, which age groups perform better or worse on average?
```

```python
data['Athlete age'] = data['Year of event'] -  data['Athlete year of birth']
```

```python
data
```

```python
data.dropna(subset=['Athlete age category'])
```

```python
# Function to convert time strings to total seconds
def time_to_seconds(time_str):
    if pd.isna(time_str):
        return None
    hours, minutes, seconds = map(int, time_str.split(':'))
    return hours * 3600 + minutes * 60 + seconds
```

```python
# Function to process average speed
def process_average_speed(speed_str):
    try:
        # Attempt to convert to float
        speed = float(speed_str)
        return speed  # Keep it as is if already a float
    except ValueError:
        # If it fails, assume it's a time string
        return time_to_seconds(speed_str)
```

```python
# Apply the processing function to the average speed column
data['Average speed (seconds)'] = data['Athlete average speed'].apply(process_average_speed)
```

```python
data
```

```python
data.dropna(subset=['Athlete age', 'Average speed (seconds)'])
```

```python
## Calculate the correlation between age and average speed in seconds
correlation = data['Athlete age'].corr(data['Average speed (seconds)'])
print(f"Correlation between athlete age and average speed: {correlation}")
```

```python
# Grouping by age category to find which age groups perform better or worse on average
age_category_performance = data.groupby('Athlete age category')['Average speed (seconds)'].mean().sort_values(ascending=True)

print("\nAverage speed by age category (in seconds):")
print(age_category_performance)
```
