```python
## Which are Top Selling course??
data.groupby('course_title').num_subscribers.sum().sort_values(ascending=False)
```

```python
data.sort_values('num_subscribers', ascending=False)
```

```python
## which are least selling courses
data.groupby('course_title').num_subscribers.sum().sort_values(ascending=True)
```
